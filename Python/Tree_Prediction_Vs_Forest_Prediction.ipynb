```python
import pandas as pd  
import numpy as np
iris_data=pd.read_csv("C:/users/govind/iris.csv")
```

```python
iris_data.head()
```

```python
iris_data.tail()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.pairplot(iris_data,hue='Species')
```

```python
X = iris_data.drop('Species',axis=1)
```

```python
X.head() 
```

```python
y = iris_data['Species']
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.4)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree_model = DecisionTreeClassifier()
tree_model.fit(X_train,y_train)
```

```python
tree_prediction = tree_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
mat=confusion_matrix(y_test,tree_prediction)
sns.heatmap(mat,square=True,annot=True,cbar=False)
print(classification_report(y_test,tree_prediction))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier(n_estimators=200)
```

```python
rf_model.fit(X_train,y_train)
```

```python
rf_model_prediction = rf_model.predict(X_test)
```

```python
mat1=confusion_matrix(y_test,rf_model_prediction)
sns.heatmap(mat1,square=True,annot=True,cbar=False)
print(classification_report(y_test,rf_model_prediction))
```
